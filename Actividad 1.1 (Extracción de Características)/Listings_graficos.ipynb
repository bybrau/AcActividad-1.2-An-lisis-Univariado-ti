```python
## importo las librerías necesarias ##
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from funpymodeling.exploratory import freq_tbl
```

```python
## Leo mi documento e imprimo las primeras dos filas ##
listings= pd.read_csv("New_Brunswick50_Sin_Atipicos1.csv")
listings.head(2)
```

```python
## Veo la información que contiene por columna ##
## Corroboro que mi documento solo cuente con 50 columnas ##
listings.info()
```

```python
## Corroboro que mi documento no tenga valores nulos ##
listings.isnull().sum()
```

```python
## Obtengo un análisis univariado de las variables categóricas ##
freq_tbl(listings)
```

```python
#Obtengo un análisis univariado de una variable categórica en específico
tabla_obj_1= freq_tbl(listings['host_location'])
#Elimino las columnas percentage y cumulative perc para trabajar con las frecuencias
tabla_obj_1= tabla_obj_1.drop(['percentage','cumulative_perc'],axis=1)
#tabla_obj_1

#Filtro de los valores más relevantes de las variables (mayores a 8)
filtro_to1_1 = tabla_obj_1 [tabla_obj_1['frequency']>8]
#Ajusto el índice de mi dataframe
filtro_to1_1 = filtro_to1_1.set_index('host_location')
#filtro_to1_1
```

```python
filtro_to1_1.plot(kind = 'bar', width = 0.8, figsize=(10,4), color="green")
plt.title('Ubicación de los registros')
plt.xlabel('Ubicación del registro')
plt.ylabel('Registros (casas registradas)')
```
